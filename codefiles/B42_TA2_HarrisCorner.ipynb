```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('corner1.jpg', cv2.IMREAD_GRAYSCALE)
img_float = np.float32(img)
dst = cv2.cornerHarris(img_float, 2, 3, 0.04)
dst_dilated = cv2.dilate(dst, None)
threshold = 0.01 * dst_dilated.max()
img_corners = img.copy()
img_corners[dst_dilated > threshold] = 255
img_corners_rgb = cv2.cvtColor(img_corners, cv2.COLOR_GRAY2RGB)
img_corners_rgb[dst_dilated > threshold] = [255, 0, 0]

plt.figure(figsize=(8, 8))
plt.title("Harris Corner Detection")
plt.imshow(img_corners_rgb)
plt.axis('off')
plt.show()
```
